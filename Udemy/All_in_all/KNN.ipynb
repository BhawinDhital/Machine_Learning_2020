```python
from sklearn.datasets import load_iris
import pandas as pd
iris  = load_iris()
df = pd.DataFrame(iris.data)

df.columns = iris.feature_names
df['Iris_Species'] = iris.target
df.isnull().values.any()

X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)

# Feature Scaling(Standardization)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# Fitting K-NN to the Training set
from sklearn.neighbors import KNeighborsClassifier
knn_c = KNeighborsClassifier(n_neighbors = 3)
knn_c.fit(X_train, y_train)
# Predicting the results on Test set
y_pred_kneigh = knn_c.predict(X_test)
# Creating the Confusion Matrix
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred_kneigh))


from sklearn.linear_model import LogisticRegression
lr_c = LogisticRegression(random_state = 0, solver='lbfgs') 
lr_c.fit(X_train, y_train)
# Predicting the results on Test set
y_pred_lr = lr_c.predict(X_test)
# Creating the Confusion Matrix
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred_lr))


from sklearn.naive_bayes import GaussianNB

# create Gaussian Naive Bayes model object and train it with the data
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)
# Predicting the results on Test set
y_pred_nb = nb_model.predict(X_test)
# Creating the Confusion Matrix
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred_nb))
```
