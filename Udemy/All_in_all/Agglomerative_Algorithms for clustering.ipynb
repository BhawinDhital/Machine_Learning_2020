### Agglomerative Algorithm for clustering
First make clustering and then apply agglomerative algorithms

```python
import pandas as pd
import matplotlib.pyplot as plt
dataset = pd.read_csv("Mall_Customers.csv")
X = dataset.iloc[:, [3,4]].values

import scipy.cluster.hierarchy as sch_clus
sch_clus.dendrogram(sch_clus.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
h_c = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')

pred_hc = h_c.fit_predict(X)

size = 25
plt.scatter(X[pred_hc == 0, 0], X[pred_hc == 0, 1], s = size, c = 'red')
plt.scatter(X[pred_hc == 1, 0], X[pred_hc == 1, 1], s = size, c = 'blue')
plt.scatter(X[pred_hc == 2, 0], X[pred_hc == 2, 1], s = size, c = 'green')
plt.scatter(X[pred_hc == 3, 0], X[pred_hc == 3, 1], s = size, c = 'yellow')
plt.scatter(X[pred_hc == 4, 0], X[pred_hc == 4, 1], s = size, c = 'pink')
plt.title('customers Cluster')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
