#### Boston Housing

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston
boston = load_boston()
df = pd.DataFrame(boston.data)
print(boston.feature_names)
```

```python
#print(df.isnull().values.any())
print(df.corr())
```

```python
import matplotlib.pyplot as plt 
def plot_corr(df):
    corr = df.corr()    
    fig, ax = plt.subplots(figsize=(10, 10))
    ax.matshow(corr)
    plt.xticks(range(len(corr.columns))) 
    plt.yticks(range(len(corr.columns))) 
plot_corr(df)

X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)

from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
#R-square training set
print("R-Square Training Set: {0}".format(linear_model.score(X_train, y_train)))
#R-square test set
print("R-Square Test Set: {0}".format(linear_model.score(X_test, y_test)))

y_predict = linear_model.predict(X_test)
```
